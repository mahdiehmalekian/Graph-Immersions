# Drawing rooted $W_4$ obstructions

We will use SageMath to draw the Type 3 and Type 4 obstructions to immersion of rooted $W_4$ (as described in DeVos, Malekian, The structure of graphs with no W_4 immersion, https://arxiv.org/abs/1810.12863)

```python
def read_graphs(file_path):
    """Reads graphs from rooted W_4 obstructions.txt file."""
    L = [] 
    with open(file_path) as file_object:
        for line in file_object:
            if not 'The connected graph with' in line:
                continue
            (root, edges_string) = line.split(' and edge set {')
            root = int(root[-1])
            all_edges = edges_string[1:-3]
            edges = []
            string_pairs = all_edges.split(", (")
            for pair in string_pairs:
                node_1 = int(pair[0])
                node_2 = int(pair[3])
                edge_mult = int(pair[-1])
                for i in range(edge_mult):
                    edges.append((node_1, node_2))            
            G = Graph(edges, multiedges=True)
            L.append((G, root))
    return L
```

```python
file_path = 'rooted W_4 obstructions.txt'

L = read_graphs(file_path)
for (G, root) in L:
    print('The following graph with root ', root)
    G.show()
```
