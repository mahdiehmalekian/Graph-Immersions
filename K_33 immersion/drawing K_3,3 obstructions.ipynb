# Drawing  $K_{3,3}$ obstructions

We will use SageMath to draw the Type 3 and Type 4 obstructions to immersion of $K_{3,3}$ (as described in DeVos, Malekian, The structure of graphs with no K_3,3 immersion, https://arxiv.org/abs/1810.12873)

```python
def read_graphs(file_path):
    """Reads graphs from K_3,3 obstructions.txt file."""
    L = [] 
    with open(file_path) as file_object:
        for line in file_object:
            if not 'The connected graph with' in line:
                continue
            edges_string = line.split(' edge set {')[1]
            all_edges = edges_string[1:-3]
            edges = []
            string_pairs = all_edges.split(", (")
            for pair in string_pairs:
                node_1 = int(pair[0])
                node_2 = int(pair[3])
                edge_mult = int(pair[-1])
                for i in range(edge_mult):
                    edges.append((node_1, node_2))            
            G = Graph(edges, multiedges=True)
            L.append(G)
    return L
```

```python
file_path = 'K_3,3 obstructions.txt'

L = read_graphs(file_path)
for G in L:
    G.show()
```
