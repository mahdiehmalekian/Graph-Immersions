# Drawing  $K_{3,3}$ obstructions

We will use SageMath to draw the'Type 3' and 'Type 4' obstructions to $K_{3,3}$.

```python
def extract_edges(string):
    """Extract graph edges from K_3,3 obstructions.txt file."""
    all_edges = string[1:-3]
    edges = []
    string_pairs = all_edges.split(", (")
    for pair in string_pairs:
        node_1 = int(pair[0])
        node_2 = int(pair[3])
        edge_mult = int(pair[-1])
        for i in range(edge_mult):
            edges.append((node_1, node_2))
    return edges
```

```python
file_path = 'K_3,3 obstructions.txt'

with open(file_path) as file_object:
    for line in file_object:
        if 'The connected graph with' in line:
            edges_string = line.split(' edge set {')[1]
            edges = extract_edges(edges_string)
            G = Graph(edges, multiedges=True)
            G.show()
```
